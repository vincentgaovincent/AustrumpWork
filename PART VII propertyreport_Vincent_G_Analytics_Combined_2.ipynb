```python
matched_df1=analytics_auctionresult_df.merge(analytics_forsale_df, left_on = "match", right_on = "match", how = "left")
matched_df1=matched_df1.drop("match",axis=1)

# create excel writer object
writer=pd.ExcelWriter(f'Matched_Suburb_for_Sale_AuctionResult{suburb}_2.xlsx')
# write dataframe to excel
matched_df1.to_excel(writer, index=False)
# save the excel
writer.save()
print('DataFrame is written successfully to Excel File.')
matched_df1.head(2)
```

```python
matched_df2=analytics_auctionresult_df.merge(analytics_forsale_df, left_on = "match", right_on = "match", how = "right")
matched_df2=matched_df2.drop("match",axis=1)

# create excel writer object
writer=pd.ExcelWriter(f'Matched_Suburb_AuctionResult_for_Sale{suburb}_2.xlsx')
# write dataframe to excel
matched_df2.to_excel(writer, index=False)
# save the excel
writer.save()
print('DataFrame is written successfully to Excel File.')
matched_df2.head(2)
```
